```python
import pandas as pd
```

```python
df = pd.read_csv('WorldSustainabilityDataset.csv')
df
```

```python
df_2018 = df[df['Year'] == 2018].copy()
print(f"There are {df_2018.shape[0]} records from 2018.\n")
```

```python
columns = [
    'Country Name',
    'Renewable energy consumption (% of total final energy consumption) - EG.FEC.RNEW.ZS',
    # 'Renewable electricity output (% of total electricity output) - EG.ELC.RNEW.ZS',
    'Annual production-based emissions of carbon dioxide (CO2), measured in million tonnes',
    'Adjusted savings: carbon dioxide damage (% of GNI) - NY.ADJ.DCO2.GN.ZS',
    'Adjusted savings: natural resources depletion (% of GNI) - NY.ADJ.DRES.GN.ZS',
]
```

```python
# keep only the columns we need
df_selected = df_2018[columns]
```

Missing values

```python
# check how many missing values
print(df_selected.isnull().sum())
```

Data cleaning

```python
# do not have to clean the country name, so started from the second column
df_cleaned = df_selected.dropna(subset=columns[1:])
```

```python
if df_cleaned.shape[0] > 100:
    print(f"After handling the missing values, {df_cleaned.shape[0]} countries remain available for analysis.\n")
else:
    print(f"There are {df_cleaned.shape[0]} countries left for analysis. Please consider whether to fill in the missing values using the mean.")
```

Save the data to a new file

```python
filename = 'cleaned_data.csv'
df_cleaned.to_csv(filename, index=False)
```
