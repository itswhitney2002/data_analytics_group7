```python
import pandas as pd
```

```python
df_clustered_2018 = pd.read_csv('clustered_data.csv')
df_clustered_2018
```

```python
df_raw = pd.read_csv('WorldSustainabilityDataset.csv')
df_raw
```

```python
cluster_mapping = df_clustered_2018[['Country Name', 'Cluster']]
countries_in_analysis = cluster_mapping['Country Name'].unique()
print(f"Successfully extracted {len(countries_in_analysis)} countries.")
```

```python
# keep only the countries which is in our clustered data
df_historical = df_raw[df_raw['Country Name'].isin(countries_in_analysis)].copy()
print(f"The filtered historical dataset contains a total of {df_historical.shape[0]} records.")
```

```python
df_final = pd.merge(df_historical, cluster_mapping, on='Country Name', how='left')
print("Preview of the columns in the final merged dataset", df_final.columns.tolist())
```

Test wether it successfully merge or not

```python
if 'Cluster' in df_final.columns:
    print("The 'Cluster' column has been successfully added.")
    
    # check if any rows have an NaN value in the 'Cluster' column
    if df_final['Cluster'].isnull().any():
        print("Warning: Empty Cluster values were found after merging. Please check the data.")
    else:
        print("All historical records have been successfully labelled with their respective clusters.")
else:
    print("Error: The 'Cluster' column does not exist after merging. Please check the merge logic.")
```

```python
import os
```

```python
output = 'final_cluster_datasets'
if not os.path.exists(output):
    os.makedirs(output)

clusters = df_final['Cluster'].dropna().unique() # Remove any NaN values that may have resulted from the merge.
clusters.sort()

for i in clusters:
    df_cluster = df_final[df_final['Cluster'] == i]
    
    output_filename = os.path.join(output, f'cluster_{int(i)}_data.csv')

    df_cluster.to_csv(output_filename, index=False)
    
    print(f"Successfully exported the data for Cluster {int(i)}")
    print(f"Including {df_cluster['Country Name'].nunique()} countries")
    print(f"The file save as {output_filename}")
```
